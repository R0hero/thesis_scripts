```python
# material = 'clean'
# material = 'glass'
# material = 'metal'
material = 'brick'

# signal = 'L1'
# signal = 'L5'
signal = 'L1+L5'

# chip_spacing = 'chip_0.5narrow_0.15'
chip_spacing = 'chip_0.8narrow_0.3'
# chip_spacing = 'chip_0.2narrow_0.1'

date = '230524_8'

path = f'data{os.sep}{signal}{os.sep}{material}{os.sep}{chip_spacing}{os.sep}'

obs_file = f'{date}.24O'
nav_file = f'{date}.24N'

obs = gr.load(path+obs_file)
nav = gr.load(path+nav_file)

sv_list = obs.sv.values

plt.figure()
for sv in sv_list:

    plt.scatter(obs.sel(sv=sv).time.values, obs.sel(sv=sv)['S1C'],color='blue',
                s=2)
    plt.scatter(obs.sel(sv=sv).time.values, obs.sel(sv=sv)['S5Q'],color='purple',
                s=2)

plt.ylabel(f'$C/N_0$ [dB-Hz]')
plt.xlabel(f'time')

plt.show()
```
